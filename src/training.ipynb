```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.3, random_state=42, shuffle=True, stratify=df['label'])
```

```python
y_train = train_set[['label']] - 1 #torch dataset expects 0 indexing
x_train = train_set.copy()
del x_train['label']

y_test = test_set[['label']] - 1
x_test = test_set.copy()
del x_test['label']
```

```python
import torch
from torch.utils.data import DataLoader, TensorDataset
import torch.nn as nn
```

```python
x_train = torch.tensor(x_train.values.astype(np.float32))
y_train = torch.tensor(y_train.values.astype(np.int32)).squeeze()
x_test = torch.tensor(x_test.values.astype(np.float32))
y_test = torch.tensor(y_test.values.astype(np.int32)).squeeze()

train_dataset = TensorDataset(x_train, y_train)
test_dataset = TensorDataset(x_test, y_test)
```

```python
class TransformDataset(torch.utils.data.Dataset):
    def __init__(self, dataset, transform=None):
        self.dataset = dataset
        self.transform=transform
    def __len__(self):
        return len(self.dataset)
    def __getitem__(self, index):
        x, y = self.dataset[index]
        if self.transform:
            x = self.transform(x)
        return x, y
```

```python
import torch
def transformations(x):
    noise = torch.randn_like(x) * 0.02
    scale = torch.empty(1, device=x.device).uniform_(0.98, 1.02)
    return (x + noise) * scale
```

```python
train_dataset = TransformDataset(train_dataset, transform=transformations)
```

```python
train_loader = DataLoader(train_dataset, 64, shuffle=True, num_workers=0)
test_loader = DataLoader(test_dataset, 64, shuffle=False, num_workers=0)
```

```python
class Classifier(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.feedforward = nn.Sequential(
            nn.BatchNorm1d(63),
            nn.Linear(in_features=63, out_features=256),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.BatchNorm1d(256),
            nn.Linear(in_features=256, out_features=128),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.BatchNorm1d(128),
            nn.Linear(in_features=128, out_features=5),
        )
    
    def forward(self, x):
        x = self.feedforward(x)
        return x
```

```python
device = 'mps' if torch.mps.is_available() else 'cpu'
```

```python
model = Classifier().to(device)
torch.compile(model)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.005, weight_decay=0.01)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)
```

```python
def get_accuracy():
    model.eval()
    with torch.no_grad():
        total_correct = 0
        for X, y in test_loader:
            X, y = X.to(device), y.to(device)
            
            output = model(X)
            total_correct += (torch.argmax(output, dim=1) == y).sum().item()
            
        test_acc = total_correct / len(test_loader.dataset)
        return test_acc
```

```python
def train(epochs:int):
    
    for epoch in range(epochs):
        model.train()
        epoch_loss = 0
        
        total_correct = 0
        
        for X, y in train_loader:
            X, y = X.to(device), y.to(device)
            
            optimizer.zero_grad()
            
            output = model(X)
            
            total_correct += (torch.argmax(output, dim=1) == y).sum().item()
            
            loss = loss_fn(output, y)
            epoch_loss += loss.item()
            loss.backward()
            
            optimizer.step()
            
        scheduler.step()
        train_acc = total_correct / len(train_loader.dataset)
        
        print(f"Epoch {epoch+1}: loss: {epoch_loss / len(train_loader)}, Train Accuracy: {train_acc}, Test Accuracy: {get_accuracy()}, LR = {optimizer.param_groups[0]['lr']}")
```

```python
train(20)
```

```python
import os
def save_model(model, filename="hand_gesture_model.pth"):
    if not os.path.exists("checkpoints"):
        os.makedirs("checkpoints")
        
    filepath = os.path.join("checkpoints", filename)
    torch.save(model.state_dict(), filepath)
    print(f"Model saved to {filepath}")

save_model(model)
```
